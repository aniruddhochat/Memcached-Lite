# Plotting Graph of Get and Set Operations corresponding to the number of calls for different storage systems.

```python
import matplotlib.pyplot as plt

noOfCalls = [100, 1000, 10000, 100000]
timing = {
    "VM Boot Disk": [0.06586241722106934, 0.30051445960998535, 3.050597906112671, 76.04047966003418],
    "GCS": [11.349967241287231,0,0,0],
    "Redis": [0.06280279159545898, 0.524960994720459, 5.177708625793457, 55.4121527671814]
}

fig, ax = plt.subplots(figsize=(10, 6))

barWidth = 0.2
index = range(len(noOfCalls))

for label, times in timing.items():
    plt.bar([i + barWidth * list(timing.keys()).index(label) for i in index], times, barWidth, label=label)

plt.xlabel("Number of Calls")
plt.ylabel("Time (seconds)")
plt.title("SET Operation Times for Different Number of Calls")
plt.xticks([i + barWidth for i in index], noOfCalls)
plt.legend()

plt.tight_layout()
plt.show()
```
